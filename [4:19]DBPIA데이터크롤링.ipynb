```python
import pandas as pd
```

```python
thesis_data = []

def crawl_data(pageNum):
    thesis_list = browser.find_elements(By.CLASS_NAME, 'thesis__link')

    detail_url_list = []
    for thesis in thesis_list:
        detail_url = thesis.get_attribute('href')
        detail_url_list.append(detail_url)

    for link in detail_url_list:
        browser.get(link)
        time.sleep(3)

        thesisDetails = browser.find_element(By.CLASS_NAME, 'thesisDetail')
        
        # 논문 제목
        thesis_title = thesisDetails.find_element(By.TAG_NAME, 'h1')
        print(thesis_title.text)

        # 저자 정보
        thesis_author = thesisDetails.find_element(By.CLASS_NAME, 'authorList')
        print(thesis_author.text)

        # 인용수
        thesis_recite = thesisDetails.find_element(By.CLASS_NAME, 'etcBtn')
        print(thesis_recite.text)

        data = {
            "논문제목" : thesis_title.text,
            "저자정보" : thesis_author.text,
            "인용수" : thesis_recite.text
        }

        thesis_data.append(data)

    # 원래 있던 페이지로 돌아가기
    browser.get(search_url)
    time.sleep(2)
    # 한 페이지에 있는 리스트 전부 다
    tag = '//*[@id="pageList"]/a[{}]'
    next_page_tag = tag.format(pageNum)
    time.sleep(3)
    more_btn = browser.find_element(By.XPATH, next_page_tag)
    more_btn.click()

    print("current page is", pageNum)

    

for i in range (1, 5):                                                                                                                        
    crawl_data(i)
    print("실행")


df = pd.DataFrame(thesis_data)
df.to_csv("DBPIA_list.csv", encoding='utf-8-sig')

```
